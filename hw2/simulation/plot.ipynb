```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
def draw(path,t):
    # Define the folder path
    folder_path = path

    # Initialize empty lists to store data from each file
    time_data = []
    energy_data = []
    labels = []
    title = t
    # Loop through each CSV file in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            # Read the CSV file into a pandas DataFrame
            file_path = os.path.join(folder_path, filename)

            temp = file_path.split("/")
            # print(temp)
            labels.append(temp[1][:-4])
            df = pd.read_csv(file_path)
            
            # Extract time and energy columns from the DataFrame
            time_data.append(df['time'])
            energy_data.append(df['energy'])

    # Create a plot with different colors for each file's energy data
    colors = ['b', 'g', 'r']
    alphas = [0.7,.7,.7]  # Adjust alpha values here
    linewidths = [2.0, 2.0, 2.0] 
    plt.figure(figsize=(10, 6))

    for i in range(len(energy_data)):
        plt.plot(time_data[i], energy_data[i], color=colors[i], label=labels[i],alpha=alphas[i],linewidth=linewidths[i])

    plt.xlabel('Time')
    plt.ylabel('Energy')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.savefig(str(path)+".png")

    # Show the plot
    # plt.show()
    # plt.savefig(str(path)+".png")
```

```python
draw("B_1","sub steps 1")
draw("B_2","sub steps 10")
draw("B_3","sub steps 100")
draw("B_4","relaxation steps 10")
# draw("B_5","sub steps 1 with error")
```
